```python
import numpy as np

data = np.load('data.npy')
target = np.load('target.npy')
```

```python
train_data = data[:1000]
train_target = target[:1000]

test_data = data[1000:]
test_target = target[1000:]
```

```python
from matplotlib import pyplot as plt

test_x_range = np.arange(1000,1209)

plt.plot(train_target,'r',label='train')
plt.plot(test_x_range, test_target, 'b',label='test')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Close Values-USD')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

model = Sequential()

model.add(LSTM(units=60, return_sequences=True, input_shape=(50,1)))
model.add(LSTM(units=60, return_sequences=True))
model.add(LSTM(units=60, return_sequences=False))
model.add(Dense(units=1, activation='linear'))

model.compile(loss='mse', optimizer='adam')

model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint_path = "models/model-{epoch:03d}.model"

checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_loss',save_best_only=True,mode='auto')
```

```python
history = model.fit(train_data, train_target, epochs=200, validation_split=0.1, callbacks=[checkpoint])
```

```python
plt.plot(history.history['loss'], 'r', label='loss')
plt.plot(history.history['val_loss'], 'b', label='val_loss')

plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```
