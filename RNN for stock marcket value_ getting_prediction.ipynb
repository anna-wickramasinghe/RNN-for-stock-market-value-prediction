```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from keras.models import load_model
from sklearn.metrics import r2_score

data = np.load('data.npy')
target = np.load('target.npy')

test_data = data[1000:]
test_target = target[1000:]
```

```python
test_target.shape
```

```python
model=load_model('models/model-092.model')
result = model.predict(test_data)

r2 = r2_score(test_target, result+0.0051)
print('r2_score: ',r2)
```

```python
plt.plot(test_target, 'r', label='actual')
plt.plot(result+0.0051, 'b', label='predicted')

plt.xlabel('date')
plt.ylabel('Close Value - USD')
plt.legend()
plt.show()
```
